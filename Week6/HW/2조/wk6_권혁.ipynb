```python
# (a) transition prob matrix P
P = np.array([[0.5, 0.5, 0, 0, 0, 0],
              [0.1, 0.5, 0.4, 0, 0, 0],
              [0, 0.2, 0.5, 0.3, 0, 0],
              [0, 0, 0.3, 0.5, 0.2, 0],
              [0, 0, 0, 0.4, 0.5, 0.1],
              [0, 0, 0, 0, 0.5, 0.5]])
```

```python
# (b) calculate P^10, P^50, P^100
P10 = P.copy()
for i in range (10):
    P10=P10.dot(P)
print(P10)
```

```python
P50 = P.copy()
for i in range (50):
    P50=P50.dot(P)
print(P50)
```

```python
P100 = P.copy()
for i in range (100):
    P100=P100.dot(P)
print(P100)
```

```python
P1000 = P.copy()
for i in range (1000):
    P1000=P1000.dot(P)
print(P1000)
```

```python
# (c) initial state of flea doesn't really matter the number of fleas t = 100
# set different initial states
Case_0 = np.array([1,0,0,0,0,0])
Case_1 = np.array([0,1,0,0,0,0])
Case_2 = np.array([0,0,1,0,0,0])
Case_3 = np.array([0,0,0,1,0,0])
Case_4 = np.array([0,0,0,0,1,0])
Case_5 = np.array([0,0,0,0,0,1])
```

```python
# t=10
print(Case_0.dot(P10))
print(Case_1.dot(P10))
print(Case_2.dot(P10))
print(Case_3.dot(P10))
print(Case_4.dot(P10))
print(Case_5.dot(P10))
# different prob. obtained
```

```python
# t=100
print(Case_0.dot(P100))
print(Case_1.dot(P100))
print(Case_2.dot(P100))
print(Case_3.dot(P100))
print(Case_4.dot(P100))
print(Case_5.dot(P100))
# same prob obtained for different initial states!!
```

```python
# (d) when t=50000, compare D and B(5,1/2) 
np.random.seed(1)
chain=0
chainHist=chain
for i in range(50000):
    pb=P[chain]
    pb=np.array(pb).flatten()
    chain=np.random.choice(6,1,p=pb)
    chainHist=np.append(chainHist,chain)

count={}
for i in chainHist:
    try: count[i] += 1
    except: count[i]=1
print(count)
plt.figure()
bi=np.random.binomial(5, 0.5, 50000)
plt.hist(bi,bins=np.linspace(-0.5,5.5,7), alpha=0.5, density=True)
plt.hist(chainHist, bins=np.linspace(-0.5,5.5,7),alpha=0.5, density=True)
plt.show()
```
