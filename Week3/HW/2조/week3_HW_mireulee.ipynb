Assume the weights in the population are normally distributed with unknown mean $\theta$ and known standared deviation 20 pounds.  Suppose your prior distribution for $\theta$ is normal with mean 180 and standard devieation 40. 

(a) give your posterior distribution for $\theta$

(c) for n = 10, give a 95% posterior interval for $\theta$ and a 95% posterior predictive interval for $\hat{y}$

```python
n = 10
posterior_mean = (600*n + 180)/(4*n + 1)
posterior_var = 1600/(4*n + 1)
posterior = scipy.stats.norm(loc = posterior_mean, scale = np.sqrt(posterior_var))
l1 = posterior.ppf(0.025)
u1 = posterior.ppf(0.975)

predictive_mean = (600*n + 180)/(4*n + 1)
predictive_var = (1600*n + 2000)/(4*n + 1)
predictive = scipy.stats.norm(loc = predictive_mean, scale = np.sqrt(predictive_var))
l2 = predictive.ppf(0.025)
u2 = predictive.ppf(0.975)

x = np.linspace(0, 200, 1000)
plt.plot(x, posterior.pdf(x), 'b-', label = 'posterior')
plt.plot(x, predictive.pdf(x),'r-', label = 'predictive')
plt.legend()
plt.show()
print("posterior 분포의 95% CI: ", l1, u1)
print("predictive 분포의 95% PI: ", l2, u2)
```
