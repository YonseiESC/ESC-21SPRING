```python
import scipy
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom
from scipy.stats import gamma

# Likelihood
# 밑에 factorial 어쩌구저쩌구는 없어도 된다. Bayesian에서 likeliood 는 분자 분모 둘다에 있고,
# 적분은 theta(모수) 에 대해서 하기 때문. 그러므로 모수와 관련없는 constant 는 다 무시 가능

def likelihood(theta, n, sy):
    return (theta**sy)*np.exp(-n*theta)

# prior for group A
aA = 120
bA = 10
priorA = st.gamma(aA, scale= 1/bA)

# prior for group B
aB = 12
bB = 1
priorB = st.gamma(aB, scale= 1/bB)

# data in group A
nA = 10
syA = 117
postA = st.gamma(aA+syA, scale=1/(bA+nA))

# data in group B
nB = 13
syB = 113
postB = st.gamma(aB+syB, scale=1/(bB+nB))
```

**Group A Bayesian Analysis**

```python
# posterior mean (Group A)
(aA+syA)/(bA+nA)
```

```python
# posterior mode (Group A)
(aA+syA-1)/(bA+nA)
```

```python
# posterior 95% CI (Group A)
# upper and lower bounds
lbA = gamma.ppf(0.025, aA+syA, scale=1/(bA+nA))
ubA = gamma.ppf(0.975, aA+syA, scale=1/(bA+nA))
(lbA, ubA)
```

```python
thetas = np.linspace(0.001, ubA+1, 300)
plt.figure(figsize=(8, 6))
plt.style.use('ggplot')
plt.plot(thetas, priorA.pdf(thetas), label='Prior', c='blue')
# 시각화를 위해 likelihood 함수에 임의의 상수를 곱함.
# plt.plot(thetas, likelihood(thetas, nA, syA), label='A Likelihood', c='orange')
plt.plot(thetas, postA.pdf(thetas), label='A Posterior', c='red')
plt.xlabel(r'$\theta$', fontsize=14)
plt.ylabel('Density', fontsize=16)
plt.legend();
```

**Group B Bayesian Analysis**

```python
# posterior mean (Group B)
(aB+syB)/(bB+nB)
```

```python
# posterior mode (Group B)
(aB+syB-1)/(bB+nB)
```

```python
# posterior 95% CI (Group B)
# upper and lower bounds
lbB = gamma.ppf(0.025, aB+syB, scale=1/(bB+nB))
ubB = gamma.ppf(0.975, aB+syB, scale=1/(bB+nB))
(lbB,ubB)
```

```python
thetas = np.linspace(0.001, ubB+1, 300)
plt.figure(figsize=(8, 6))
plt.style.use('ggplot')
plt.plot(thetas, priorB.pdf(thetas), label='Prior', c='blue')
# 시각화를 위해 likelihood 함수에 임의의 상수를 곱함.
# plt.plot(thetas, (10**17)*likelihood(thetas, nB, syB), label='B Likelihood', c='orange')
plt.plot(thetas, postB.pdf(thetas), label='B Posterior', color='green')

plt.xlabel(r'$\theta$', fontsize=14)
plt.ylabel('Density', fontsize=16)
plt.legend();
```

**Comparing Two Posterior Distributions**

```python
thetas = np.linspace(0.001, ubA+1, 300)
plt.figure(figsize=(8, 6))
plt.style.use('ggplot')
plt.plot(thetas, priorA.pdf(thetas), label='A Prior', c='blue')
plt.plot(thetas, priorB.pdf(thetas), label='B Prior', c='orange')
plt.plot(thetas, postA.pdf(thetas), label='A Posterior', c='red')
plt.plot(thetas, postB.pdf(thetas), label='B Posterior', c='green')
plt.xlabel(r'$\theta$', fontsize=14)
plt.ylabel(r'$Pr(\theta|data)$', fontsize=16)
plt.legend();
```

**Posterior Predictive Distributions**

```python
ynew = np.arange(30)
postpredA = st.nbinom.pmf(ynew, aA+syA, (bA+nA)/(bA+nA+1))
postpredB = st.nbinom.pmf(ynew, aB+syB, (bB+nB)/(bB+nB+1))
width = 0.35  # the width of the bars
plt.style.use('ggplot')
plt.figure(figsize=(8, 6))
plt.bar(ynew - width/2, postpredA, width, label='group A', color='red', alpha=0.5)
plt.bar(ynew + width/2, postpredB, width, label='group B', color='blue', alpha=0.5)
plt.xlabel('Ynew')
plt.ylabel(r'$Pr(Ynew|data)$')
plt.legend()
plt.show()
```

### b) n0가 커야 함 (near 50). 결론적으로 strong prior belief 필요

```python
plt.style.use('ggplot')
plt.figure(figsize=(8, 6))
for n0 in range(1,51):
    exp = (aB*n0+syB)/(bB*n0+nB)
    plt.scatter(n0, exp, color='blue')
```

**왜 Beta distribution이 Binomial data(likelihood)의 conjugate prior가 될 수 있을까?**  
  
\- 우선 parameter space가 같고,  
\- exponential family로 변환했을 때 $\theta^ㅁ (1-\theta)^ㅇ$로 format도 비슷하다.  
\- parameter 업데이트를 의미적으로, 직관적으로 해석하기도 가능해짐  
  
prior를 $p(\theta) \sim Beta(n_0t_0, n_0(1-t_0))$로 줬을 때, $E[\theta] = \dfrac{n_0t_0}{n_0} = t_0$가 된다.  
이 때, $t_0$를 prior guess라고 자연스럽게(?) 받아들일 수 있다...!

## Relationship between Poisson distribution and Neg-Binom distribution
### Prove the following.
$$Poi(\lambda) = \lim_{r \rightarrow \infty} NB\Big(r, \dfrac{\lambda}{r + \lambda}\Big)$$
  
### Compare the variance of each distribution. Show that the Neg-Binom distribution is always overdispersed.  
  
아래에서 보이는 바와 같이 Poisson distribution은 Neg-Binomial distribution에 비해 overdispersion problem이 존재한다. 나도 헷갈렸던 건데, 여기서 overdispersion이란 Poisson 모델에서는 평균=분산이기 때문에 실제로 분산이 큰 데이터를 모델링하지 못하는 경우를 말한다. (**In statistics, overdispersion is the presence of greater variability (statistical dispersion) in a data set than would be expected based on a given statistical model.**) 따라서, 이러한 경우 additional free variable을 도입하여 Neg-Binomial model을 채택할 수 있다. Poisson model의 parameter는 하나인데 반해, Poisson을 극한분포로 갖는 Neg-Binom model은 두 개의 parameter를 갖기 때문에 좀 더 유동적으로 분산이 큰 데이터를 모델링할 수 있다.
  
### Probe the following
Let mean $np=\lambda \ \ \rightarrow \ \ p = \dfrac{\lambda}{n}$
$$Poi(\lambda) = \lim_{r \rightarrow \infty} Binom\Big(n, \dfrac{\lambda}{n}\Big)$$
