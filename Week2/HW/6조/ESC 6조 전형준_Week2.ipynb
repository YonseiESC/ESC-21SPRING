a) Find the posterior distributions, means, variances and 95% quantilebased
confidence intervals for A and B, assuming a Poisson sampling
distribution for each group and the following prior distribution:
A  gamma(120,10), B  gamma(12,1), p(A, B) = p(A)×p(B).

```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom
from scipy.stats import gamma

ya = [12, 9, 12, 14, 13, 13, 15, 8, 15, 6]
yb = [11, 11, 10, 9, 9, 8, 7, 10, 6, 8, 8, 9, 7]
na = len(ya); nb = len(yb)
sa = sum(ya); sb = sum(yb)
al_a = 120; be_a = 10; al_b=12; be_b=1
postmean_a = (al_a+sa)/(be_a+na)
postvar_a = (al_a+sa)/(be_a+na)**2
lba = gamma.ppf(0.025, al_a+sa, scale=1/(be_a+na))
uba = gamma.ppf(0.975, al_a+sa, scale=1/(be_a+na))
print("Posterior distribution of a is gammma(%d, %d)" %(al_a+sa,be_a+na))
print("Posterior mean of a =",postmean_a)
print("Posterior variance of a =",postvar_a)
print("95% CI of a =",(lba, uba))
postmean_b = (al_b+sb)/(be_b+nb)
postvar_b = (al_b+sb)/(be_b+nb)**2
lbb = gamma.ppf(0.025, al_b+sb, scale=1/(be_b+nb))
ubb = gamma.ppf(0.975, al_b+sb, scale=1/(be_b+nb))
print("Posterior distribution of b is gammma(%d, %d)" %(al_b+sb,be_b+nb))
print("Posterior mean of b =",postmean_b)
print("Posterior variance of b =",postvar_b)
print("95% CI of b =",(lbb, ubb))
```
