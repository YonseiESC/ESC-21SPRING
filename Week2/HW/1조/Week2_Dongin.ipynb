```python
import scipy
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom
from scipy.stats import gamma
```

```python
# Likelihood
def likelihood(theta, n, sy):
    return (theta**sy)*np.exp(-n*theta)
```

```python
y_A=[12,9,12,14,13,13,15,8,15,6]
y_B=[11,11,10,9,9,8,7,10,6,8,8,9,7]
```

1-a)

```python
sum_A=sum(y_A); n_A=len(y_A); alpha_A=120; beta_A=10
sum_B=sum(y_B); n_B=len(y_B); alpha_B=12; beta_B=1
```

```python
print('Posterior mean of group A:', (alpha_A+sum_A)/(beta_A+n_A))
print('Posterior variance of group A:', (alpha_A+sum_A)/((beta_A+n_A)**2))
print('Posterior mean of group B:', (alpha_B+sum_B)/(beta_B+n_B))
print('Posterior variance of group B:', (alpha_B+sum_B)/((beta_B+n_B)**2))
```

```python
print('95% quantile baesd confidence interval of group A', 
      (round(gamma.ppf(0.025, alpha_A+sum_A, scale=1/(beta_A+n_A)),4), 
       round(gamma.ppf(0.975, alpha_A+sum_A, scale=1/(beta_A+n_A)),4)))
print('95% quantile baesd confidence interval of group B', 
      (round(gamma.ppf(0.025, alpha_B+sum_B, scale=1/(beta_B+n_B)),4), 
       round(gamma.ppf(0.975, alpha_B+sum_B, scale=1/(beta_B+n_B)),4)))
```

```python
thetas = np.linspace(0.001, 20, 300)
plt.figure(figsize=(8, 6))
plt.style.use('ggplot')
plt.plot(thetas, st.gamma(alpha_A+sum_A, scale=1/(beta_A+n_A)).pdf(thetas), label='Posterior of group A', color='red')
plt.plot(thetas, st.gamma(alpha_B+sum_B, scale=1/(beta_B+n_B)).pdf(thetas), label='Posterior of group B', color='blue')
plt.legend()
```

1-b)

```python
posterior_expectation=[]
for i in range(1,51):
    posterior_expectation.append((12*i+sum_B)/(i+n_B))
```

```python
n=np.linspace(1,50,50)
plt.figure(figsize=(8, 6))
plt.style.use('ggplot')
plt.plot(n, posterior_expectation, marker='o',label='Posterior Expectation', c='black')
plt.legend()
```
